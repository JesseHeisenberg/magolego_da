# Постановка статистической задачи и теоретические сведения

Цель исследования: проверка наличия статистиечскиой связи между полом человека и фактом курения: влияет ли пол на приверженность к курению. Оба принака измеряются в номинальной шкале. 

В рамках представленного набора данных я рассматриваю две величины

* sex — пол человека (Male/Female);
* SMK_stat_type_cd — статус курения (Y/N).

Для этого сформулировал следующие гипотезы:

$$
H_0: X \text{ и } Y \text{ независимы, т.е. вероятность принадлежности человека к группе курящих одинакова для мужчин и женщин} 
$$

$$
H_1: X \text{ и } Y \text{ гипотеза предполагает существование статистической зависимости между рассматриваемыми признаками}
$$

---

## Таблица сопряжённости

Для анализа двух номинальных признаков строится таблица сопряжённости, содержащая частоты наблюдений для всех возможных комбинаций значений признаков.

Общий вид таблицы:

|       | $$Y_1$$        | $$Y_2$$        | $$...$$ | $$Итого$$        |
| ----- | ------------ | ------------ | --- | ------------ |
| $$X_1$$ | $$O_{11}$$     | $$O_{12}$$     | $$...$$ | $$O_{1\cdot}$$ |
| $$X_2$$ | $$O_{21}$$     | $$O_{22}$$    | $$...$$ | $$O_{2\cdot}$$ |
| $$...$$   | $$...$$          | ...          | $$...$$ | $$...$$          |
| $$Итого$$ | $$O_{\cdot1}$$ | $$O_{\cdot2}$$ | $$...$$ | $$n$$          |

Где в таблице: 
$$
O_{ij} — наблюдаемая частота;
n — объём выборки.
$$

Если признаки независимы, то частоты в таблице должны быть близки к тем, которые ожидаются при независимом распределении признаков.

---

## Критерий хи-квадрат Пирсона

Для проверки гипотезы о независимости используется критерий хи-квадрат Пирсона, благодаря которому сначала вычисляются ожидаемые частоты:

$$
E_{ij}=
\frac{O_{i\cdot}\cdot O_{\cdot j}}{n}
$$

где:
$$
O_{i\cdot} — сумма наблюдений в (i)-й строке;
O_{\cdot j} — сумма наблюдений в (j)-м столбце;
n — общий объём выборки.
$$

Затем вычисляется статистика критерия:
$$
\chi^2=
\sum_{i=1}^{r}
\sum_{j=1}^{c}
\frac{(O_{ij}-E_{ij})^2}{E_{ij}}
$$

где:

* r — число строк таблицы сопряжённости;
* c — число столбцов таблицы сопряжённости.

Чем сильнее наблюдаемые частоты отличаются от ожидаемых, тем больше значение статистики **chi^2**, а следовательно, тем больше оснований считать признаки зависимыми.

При выполнении нулевой гипотезы статистика имеет распределение хи-квадрат с числом степеней свободы

$$
df=(r-1)(c-1)
$$

Для таблицы (2 \times 2):

$$
df=(2-1)(2-1)=1
$$

Для принятия решения задаю уровень значимости
$$
\alpha = 0.05.
$$

---

## Меры связи для номинальных признаков

Сам факт отклонения гипотезы ещё не показывает силу связи между признаками. Поэтому дополнительно рассчитываются специальные коэффициенты.

### Коэффициент Фи

Для таблиц размера 2X2 используется коэффициент Фи:

$$
\phi=
\sqrt{\frac{\chi^2}{n}}
$$

где:

* (\chi^2) — значение статистики Пирсона;
* (n) — объём выборки.

Коэффициент принимает значения от 0 до 1:

* 0 — связь отсутствует;
* 1 — максимально сильная связь.

---

### Коэффициент Крамера

Для таблиц произвольного размера применяется коэффициент V Крамера:

$$
V=
\sqrt{
\frac{\chi^2}
{n\cdot\min(r-1,c-1)}
}
$$

Коэффициент также изменяется в диапазоне от 0 до 1 и позволяет оценить тесноту связи между качественными признаками.

---

## Интерпретация силы связи

Для коэффициента Фи и коэффициента V Крамера часто используют следующую шкалу интерпретации:

| Значение коэффициента | Сила связи              |
| --------------------- | ----------------------- |
| 0 – 0.10              | практически отсутствует |
| 0.10 – 0.30           | слабая                  |
| 0.30 – 0.50           | умеренная               |
| более 0.50            | сильная                 |

Таким образом, в данной работе сначала будет выполнена проверка гипотезы о независимости пола и факта курения с помощью критерия хи-квадрат Пирсона, после чего сила обнаруженной связи будет оценена с использованием коэффициента Фи и коэффициента V Крамера. Это позволит не только определить наличие статистической зависимости, но и оценить её практическую значимость.


```python
# Построение таблицы сопряженности 
df_for_table = pd.concat([group_female, group_male])
table = pd.crosstab(df_for_table['sex'], df_for_table['SMK_stat_type_cd'], margins=True)

print("Таблица сопряженности: \n")
print(table)
```

```python
# Наблюдаемые частоты  
O = np.array([
    [164681, 197706],
    [437760, 16248]
])

n = O.sum()
row_totals = O.sum(axis=1)
col_totals = O.sum(axis=0)
print("Объем выборки:", n)

# Ожидаемые частоты
E = np.zeros(O.shape)

for i in range(O.shape[0]):
    for j in range(O.shape[1]):
        E[i, j] = row_totals[i] * col_totals[j] / n

print("\nОжидаемые частоты:")
print(E)

# Расчет χ² 
chi_square = 0

for i in range(O.shape[0]):
    for j in range(O.shape[1]):
        term = (O[i, j] - E[i, j])**2 / E[i, j]
        chi_square += term

        print(
            f"Ячейка ({i+1},{j+1}): "
            f"({O[i,j]} - {E[i,j]:.2f})² / {E[i,j]:.2f} "
            f"= {term:.4f}"
        )
print(f"\nχ² = {chi_square:.4f}")

r, c = O.shape
df = (r - 1) * (c - 1)
print(f"Степени свободы = {df}")

alpha = 0.05
chi_crit = chi2.ppf(1 - alpha, df)
print(f"Критическое значение χ² = {chi_crit:.4f}")

p_value = 1 - chi2.cdf(chi_square, df)
print(f"p-value = {p_value:.10f}")

phi = np.sqrt(chi_square / n)
print(f"\nКоэффициент Фи = {phi:.4f}")

cramers_v = np.sqrt(
    chi_square / (n * min(r - 1, c - 1))
)
print(f"Крамера = {cramers_v:.4f} \n")

print('='*10)

if chi_square > chi_crit:
    print("\nОтвергаем H0.")
    print("Пол и факт курения статистически зависимы.")
else:
    print("\nНет оснований отвергать H0.")
    print("Статистически значимая связь не обнаружена.")
```

Одной из причин может быть различие в социальных нормах и моделях поведения мужчин и женщин. Мужчины чаще вовлекаются в практики, связанные с употреблением табака, а также подвергаются большему влиянию соответствующих социальных факторов. Для женщин, напротив, курение во многих обществах встречается реже вследствие культурных ограничений и более негативного отношения со стороны общества.

Следует отметить, что очень большое значение статистики χ² связано не только с реальным различием между группами, но и с чрезвычайно большим объёмом выборки: **n=816395**

При таком количестве наблюдений даже относительно небольшие различия между группами становятся статистически значимыми. Поэтому наряду с проверкой гипотезы важно анализировать меры силы связи. В данном исследовании коэффициенты Фи и Крамера также показали достаточно сильную связь, что подтверждает практическую значимость полученного результата.
Таким образом, проведённый анализ позволяет заключить, что в рассматриваемой выборке вероятность того, что человек является курильщиком, существенно связана с его полом, причём среди мужчин курение встречается значительно чаще, чем среди женщин.
